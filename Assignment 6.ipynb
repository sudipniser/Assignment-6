Q1: We need to evaluate the integral and compare its values with the analytical value given below
------------------------------
$$\int_1^4 \sqrt{1+\frac{1}{x}} dx=3.6201842807857516612773260285461673526889575187414495235948736506...$$

```python
f1=lambda x:math.sqrt(1+1/x)
print('N|Midpoint|Trapezoid|Simpson')
print('---------------------------------')
for i in [8,16,24]:
    print(i,'|',inte.num_int(f1,1,4,i,'midpoint'),'|',inte.num_int(f1,1,4,i,'trapezoidal'),'|',inte.num_int(f1,1,4,i,'simpson'))
```

Q2:We need to integrate the following function
-----------------------------
$f_2(x)=x\sqrt{1+x}\\
f_2''(x)=\frac{3x+4}{4(1+x)^{3/2}}\quad |f_2''(x)|_{max}=1\\
f_2''''(x)=\frac{3(3x+8)}{16(1+x)^{7/2}}\quad |f_2''''(x)|_{max}=1.5$
Thus, to achieve the minimum error of 0.001 for various cases the iterations required are as follows,\
Midpoint: N = 7\
Trapezoidal: N=10\
Simpsons: N=2\
Also,
$$\int_0^1 x\sqrt{1+x} dx=0.643790283299492$$

```python
f2=lambda x: x*math.sqrt(x+1)
print('The value of the above integral using simpsons method in 7 interations is:',inte.num_int(f2,0,1,7,'midpoint'))
print('The value of the above integral using simpsons method in 10 interations is:',inte.num_int(f2,0,1,10,'trapezoidal'))
print('The value of the above integral using simpsons method in 2 interations is:',inte.num_int(f2,0,1,2,'simpson'))
```

Q3: We use the following formula to calculate the value of $\pi$
------------------------------------------------------------
$$\int_0^1\frac{4}{1+x^2}=\pi$$

```python
f3=lambda x: 4/(1+x**2)
I=[]
for N in range(10,10000,10):
    I.append(inte.MonteCarloInt(f3,0,1,N)[0])
```

```python
plt.plot(I)
plt.xlabel('Iteration number N')
plt.ylabel('Estimator $\mathcal{F}_N$')
plt.show()
```

Q4: We find the COM as follows,
-------------------
$$x_{com}=\frac{\int xdm}{\int dm}=\frac{\int_0^2 x\lambda(x)dx}{\int_0^2 \lambda(x)dx}=\frac{\int_0^2 x^3dx}{\int_0^2 x^2dx}=1.5$$

```python
X_COM_M=[inte.num_int(lambda x:x**3,0,2,i,'midpoint')/inte.num_int(lambda x:x**2,0,2,i,'midpoint') for i in range(10,1000,10)]
X_COM_T=[inte.num_int(lambda x:x**3,0,2,i,'trapezoidal')/inte.num_int(lambda x:x**2,0,2,i,'trapezoidal') for i in range(10,1000,10)]
X_COM_S=[inte.num_int(lambda x:x**3,0,2,i,'simpson')/inte.num_int(lambda x:x**2,0,2,i,'simpson') for i in range(10,1000,10)]
print('The value of center of mass is ',X_COM_S[-1])
N=[i for i in range(10,1000,10)]
plt.plot(N,X_COM_M,label='Midpoint')
plt.plot(N,X_COM_T,label='Trapezoidal')
plt.plot(N,X_COM_S,label='Simpsons')
plt.ylabel('Centre of mass (in m)')
plt.xlabel('Number of iteration N')
plt.legend()
plt.show()
```
